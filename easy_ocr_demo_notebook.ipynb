```
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

```
reader = easyocr.Reader(['en'])
```

```
img = Image('pancard.jpg')
```

```
img
```

```
output = reader.readtext('pancard.jpg')
```

```
output
```

```
pan = ""
index=0
for i in range(len(output)):
  if output[i][1]=='Permanent Account Number Card':
    pan = output[i+1][1]
    index = i+1
    break
```

```
birth = ""
index1=0
for i in range(len(output)):
  k = output[i][1]
  if "Date" in k:
    birth = output[i+1][1]
    index1 = i+1
    break
```

```
print("PAN : ",pan)
```

```
print("DATE OF BIRTH",birth)
```

**BOUNDING BOX ON IMAGE**

```
cord = output[index][0]
cord1 = output[index1][0]
```

```
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
```

```
x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
```

```
x1_min, y1_min = [int(min(idx)) for idx in zip(*cord1)]
```

```
x1_max, y1_max = [int(max(idx)) for idx in zip(*cord1)]
```

```
image = cv2.imread('pancard.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```
image = cv2.imread('pancard.jpg')
cv2.rectangle(image,(x1_min,y1_min),(x1_max,y1_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
